```python
%%sql
select
  sal.customerkey,
  sal.orderdate,
  prod.productname,
  cust.surname,
  cust.givenname,
  cust.countryfull,
  cust.continent,
  sal.quantity * netprice as net_revenue,
case
  when netprice * quantity <= 1000 then 'Low'
  when netprice* quantity > 1000 then 'High'
  end as revenue_category
from sales sal
left join customer cust
on sal.customerkey = cust.customerkey
left join product prod
on sal.productkey = prod.productkey
```

```python
## COUNT DISTINCT

%%sql
SELECT
  sal.orderdate,
  count(distinct case when cust.continent = 'Europe' then sal.customerkey end) as eu_customers,
  count(distinct case when cust.continent = 'North America' then sal.customerkey end) as na_customers,
  count(distinct case when cust.continent = 'Australia' then sal.customerkey end) as aus_customers
from sales sal
left join customer cust
on sal.customerkey = cust.customerkey
where sal.orderdate between '2023-01-01' and '2023-12-31'
group by sal.orderdate
order by sal.orderdate
```

```python
## sum with case

%%sql
SELECT
  prod.categoryname,
  sum(case when sal.orderdate between '2022-01-01' and '2022-12-31' then sal.quantity * sal.netprice * sal.exchangerate end) net_revenue_2022,
  sum(case when sal.orderdate between '2023-01-01' and '2023-12-31' then sal.quantity * sal.netprice * sal.exchangerate end) net_revenue_2023,
  sum(case when sal.orderdate between '2024-01-01' and '2024-12-31' then sal.quantity * sal.netprice * sal.exchangerate end) net_revenue_2024
from sales sal
left join product prod
on sal.productkey = prod.productkey
group by prod.categoryname
order by prod.categoryname
```

```python
## STATISTICS (MAX, MIN, MEAN, AVG, MEDIAN)
## AVG
%%sql
SELECT
  prod.categoryname,
  MAX(case when sal.orderdate between '2022-01-01' and '2022-12-31' then sal.quantity * sal.netprice * sal.exchangerate end) max_revenue_2022,
  MAX(case when sal.orderdate between '2023-01-01' and '2023-12-31' then sal.quantity * sal.netprice * sal.exchangerate end) max_revenue_2023,
  min(case when sal.orderdate between '2022-01-01' and '2022-12-31' then sal.quantity * sal.netprice * sal.exchangerate end) min_revenue_2022,
  min(case when sal.orderdate between '2023-01-01' and '2023-12-31' then sal.quantity * sal.netprice * sal.exchangerate end) min_revenue_2023,
  avg(case when sal.orderdate between '2022-01-01' and '2022-12-31' then sal.quantity * sal.netprice * sal.exchangerate end) net_avg_revenue_2022,
  avg(case when sal.orderdate between '2023-01-01' and '2023-12-31' then sal.quantity * sal.netprice * sal.exchangerate end) net_avg_revenue_2023
from sales sal
left join product prod
on sal.productkey = prod.productkey
group by prod.categoryname
order by prod.categoryname

```

```python
## median
%%sql
select
  PERCENTILE_CONT(0.5) within group (order by netprice) as median_price
from sales
```

```python
## MEDIAN

%%sql
SELECT
  prod.categoryname as category,
  PERCENTILE_CONT(0.5) within group (order by
  ( case
    when orderdate between '2022-01-01' and '2022-12-01' then (sal.quantity*sal.netprice*sal.exchangerate) end)) as  median_revenue_2022,
     PERCENTILE_CONT(0.5) within group (order by
  ( case
    when orderdate between '2023-01-01' and '2023-12-01' then (sal.quantity*sal.netprice*sal.exchangerate) end)) as  median_revenue_2023
from sales sal
left join product prod
on sal.productkey = prod.productkey
group by prod.categoryname
order by prod.categoryname
```

```python
%%sql
SELECT
  prod.categoryname as category,
  sum(case when (sal.quantity*sal.netprice*sal.exchangerate) < 398
      and sal.orderdate between  '2022-01-01' and '2022-12-31'
    then (sal.quantity*sal.netprice*sal.exchangerate) end ) as low_net_revenue_2022,
  sum(case when (sal.quantity*sal.netprice*sal.exchangerate) >= 398
       and sal.orderdate between  '2022-01-01' and '2022-12-31'
   then (sal.quantity*sal.netprice*sal.exchangerate) end ) as high_net_revenue_2022,
     sum(case when (sal.quantity*sal.netprice*sal.exchangerate) < 398
      and sal.orderdate between  '2023-01-01' and '2023-12-31'
    then (sal.quantity*sal.netprice*sal.exchangerate) end ) as low_net_revenue_2022,
  sum(case when (sal.quantity*sal.netprice*sal.exchangerate) >+ 398
       and sal.orderdate between  '2023-01-01' and '2023-12-31'
   then (sal.quantity*sal.netprice*sal.exchangerate) end ) as high_net_revenue_2022
from sales sal
left join product prod
on sal.productkey = prod.productkey
group by prod.categoryname
order by prod.categoryname
```

```python
%%sql
with percentiles as
(
SELECT

  PERCENTILE_CONT(0.25) within group (order by (sal.quantity * sal.netprice * sal.exchangerate)) as revenue_25th_percentile ,
  PERCENTILE_CONT(0.50) within group (order by (sal.quantity * sal.netprice * sal.exchangerate)) as revenue_50th_percentile ,
  PERCENTILE_CONT(0.75) within group (order by (sal.quantity * sal.netprice * sal.exchangerate)) as revenue_75th_percentile
from sales sal
where orderdate between '2022-01-01' and '2023-12-31'
)
select
 prod.categoryname as category,
 case
     when (sal.quantity * sal.netprice * sal.exchangerate) <= perc.revenue_25th_percentile then 'low'
     when (sal.quantity * sal.netprice * sal.exchangerate) >= perc.revenue_50th_percentile then 'middium'
     else 'high'
     end as revenue_category  ,
sum(sal.quantity * sal.netprice * sal.exchangerate) as net_revenue
from sales sal
left join product prod
on sal.productkey = prod.productkey,
percentiles perc
group by prod.categoryname,
         revenue_category
order by prod.categoryname


```
